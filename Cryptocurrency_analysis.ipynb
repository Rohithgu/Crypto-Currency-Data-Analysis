```python
import numpy as np
import pandas as pd
```

```python
hist=pd.read_csv(r"E:\excel files\crypto_historical_365days.csv")
mont=pd.read_csv(r"E:\excel files\crypto_monthly_summary.csv")
yearly=pd.read_csv(r"E:\excel files\crypto_yearly_performance.csv")
```

```python
crypto=pd.merge(left=hist,
                right=yearly,
                left_on=['coin_id'],
                right_on=['coin_id'],
                how="left")
                
```

```python
crypto_currency=pd.merge(left=crypto,
                         right=mont,
                         left_on=['month'],
                         right_on=['month'],
                         how="left")
```

```python
crypto_currency['month'] = pd.to_datetime(
    crypto_currency['month'],
    format="%Y-%m"
)
```

```python
crypto_currency['only_month'] = pd.to_datetime(
    crypto_currency['month'],
    format="%Y-%m"
).dt.month
```

```python
crypto_currency['only_year'] = pd.to_datetime(
    crypto_currency['month'],
    format="%Y-%m"
).dt.year
```

```python
Total_revenue_month=crypto_currency.groupby(by=["only_month"]).agg(Total_revenue=("end_price","mean"))
```

```python
Total_revenue_month
```

```python
Total_revenue_year=crypto_currency.groupby(by=["only_year"]).agg(Total_revenue=("end_price","mean"))
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
plt.figure()
sns.barplot(data=Total_revenue_month,x="only_month",y="Total_revenue")
plt.title("Month_wise Revenue")
plt.show()
```

```python
crypto_currency['volatility_7d'] = (
    crypto_currency.groupby('coin_id')['daily_return']
      .rolling(window=7)
      .std()
      .reset_index(level=0, drop=True)
)
```

```python
crypto_currency['volatility_7d'] = crypto_currency['volatility_7d'].fillna(0)
```

```python
top_price = (
    crypto_currency.groupby('coin_name_x')['price']
      .mean()
      .sort_values(ascending=False)
      .head(10)
)

top_price.plot(kind='bar', figsize=(10,5))
plt.title("Top 10 Coins by Average Price")
plt.ylabel("Average Price")
plt.show()
```

```python
sns.histplot(crypto_currency['price'], bins=50, kde=True)
plt.title("Price Distribution")
plt.show()
```

```python
daily_avg_price = crypto_currency.groupby('date')['price'].mean()

plt.figure(figsize=(12,5))
plt.plot(daily_avg_price)
plt.title("Average Price Over Time")
plt.xlabel("Date")
plt.ylabel("Price")
plt.show()
```

```python
top_volume = (
    crypto_currency.groupby('coin_name_x')['volume']
      .sum()
      .sort_values(ascending=False)
      .head(10)
)

top_volume.plot(kind='bar', figsize=(10,5))
plt.title("Top 10 Coins by Trading Volume")
plt.ylabel("Total Volume")
plt.show()
```

```python
sns.scatterplot(data=crypto_currency, x='volume', y='price', alpha=0.4)
plt.title("Price vs Volume")
plt.show()
```

```python
sns.histplot(crypto_currency['daily_return'], bins=50, kde=True)
plt.title("Daily Return Distribution")
plt.show()
```

```python
btc = crypto_currency[crypto_currency['coin_name_x'] == 'Bitcoin']

plt.figure(figsize=(12,5))
plt.plot(btc['date'], btc['price'])
plt.title("Bitcoin Price Over Time")
plt.show()
```

```python
plt.figure(figsize=(12,5))
plt.plot(btc['date'], btc['price'], label='Price')
plt.plot(btc['date'], btc['price_ma7'], label='MA7')
plt.plot(btc['date'], btc['price_ma30'], label='MA30')
plt.legend()
plt.title("Price vs Moving Averages (Bitcoin)")
plt.show()
```

```python
corr = crypto_currency[['price','volume','daily_return','volatility_7d','market_cap']].corr()

sns.heatmap(corr, annot=True, cmap='coolwarm')
plt.title("Correlation Heatmap")
plt.show()
```
